```python
import requests
import pandas as pd
import time

# Your personal access token
access_token = 'github_pat_11ASW4YNA0p8eim5cwKPet_E9uymRdHGNoGdLPZUCDTWRQ6W8xcNE'
headers = {'Authorization': f'token {access_token}'} if access_token else {}

org_urls = {
    'Microsoft': ['microsoft', 'Azure', 'dotnet', 'xamarin', 'PowerShell'],
    'Google': ['google', 'tensorflow', 'kubernetes'],
    'Meta': ['facebook', 'facebook/react'],
    'IBM': ['IBM', 'IBM/pytorch-seq2seq'],
    'Oracle': ['oracle'],
    'Salesforce': ['salesforce'],
    'SAP': ['SAP'],
    'Apple': ['apple', 'apple-oss-distributions', 'apple-open-source'],
    'Nvidia': ['NVIDIA'],
    'Tencent': ['tencent'],
    'Cisco': ['cisco'],
    'Sony': ['sony'],
    'Dell': ['dell'],
    'Intel': ['intel'],
    'Huawei': ['huawei'],
    'Samsung': ['Samsung'],
    'Broadcom': ['Broadcom'],
    'Qualcomm': ['qualcomm'],
    'Accenture': ['accenture'],
    'LG': ['lge-oss'],
    'Panasonic': ['panasonic-corporation', 'PanasonicConnect'],
    'Lenovo': ['lenovo'],
    'HP': ['HPInc'],
    'Uber': ['uber'],
    'Xiaomi': ['xiaomi'],
    'ChinaMobile': ['cmri'],
    'Verizon': ['verizon'],
    'ATT': ['att', 'attiot', 'attdevsupport', 'att-comdev', 'att-cloudnative-labs'],
    'Comcast': ['comcast', 'ComcastSamples'],
    'DeutscheTelekom': ['telekom', 'telekom-security', 'telekom-mms', 'caas-team', 'flowground', 'pan-net-security'],
    'NTT': ['nttcom'],
    'Nokia': ['nokia'],
    'Vodafone': ['vodafone'],
    'Orange': ['orange-opensource'],
    'SoftBank': ['softbankrobotics-labs', 'softbankrobotics-research'],
    'KDDI': ['kddi-research-wot']
}

prop_list = ['html_url', 'fork', 'created_at', 'updated_at', 'pushed_at', 'git_url', 'size', 'stargazers_count','watchers_count','language', 'forks_count','archived','disabled', 'open_issues_count', 'license', 'allow_forking']

# List to store repository links
all_org_repo = []

for company, org_list in org_urls.items():
    for org in org_list:
        print(f"{company} - {org}")
        
        # Handle specific repository cases like 'facebook/react'
        if '/' in org:
            url = f'https://api.github.com/repos/{org}'
            is_specific_repo = True
        else:
            url = f'https://api.github.com/orgs/{org}/repos'
            is_specific_repo = False
        
        # Parameters for pagination
        params = {'per_page': 100, 'page': 1}
        repository_links = []
        
        try:
            while True:  # Limit to 50 repos per organization
                response = requests.get(url, headers=headers, params=params)
                
                # Handle rate limiting
                if response.status_code == 403 and 'rate limit exceeded' in response.text.lower():
                    reset_time = int(response.headers.get('X-RateLimit-Reset', 0))
                    wait_time = max(reset_time - time.time(), 0) + 5
                    print(f"Rate limit exceeded. Waiting {wait_time:.0f} seconds...")
                    time.sleep(wait_time)
                    continue
                
                if response.status_code != 200:
                    print(f"Error fetching repositories for {org}: {response.status_code}")
                    break
                
                repositories = response.json()
                
                #print(len(repositories))
                
                if not repositories:
                    break
                
                # If it's a specific repository, wrap it in a list
                if is_specific_repo:
                    repositories = [repositories]
                    
                for repo in repositories:
                    repo_properties = []
                    for prop in prop_list:
                        try:
                            prop_text = repo.get(prop)
                            if prop == "license" and prop_text is not None:
                                prop_text = repo[prop]['name']
                            repo_properties.append(prop_text)
                        except Exception as e:
                            print(f"Error with property {prop}: {e}")
                            repo_properties.append(None)
                    
                    repository_links.append([company] + repo_properties)
                
                # If we're dealing with a specific repo, we only need one request
                if is_specific_repo:
                    break
                
                # Move to the next page for organization repos
                params['page'] += 1
            
            if repository_links:
                # Create a temp DataFrame to display progress
                df_temp = pd.DataFrame(repository_links, columns=['Organization'] + prop_list)
                print(f"Found {len(repository_links)} repositories for {org}")
                #print(df_temp.head(3))  # Show first 3 repos
                
                # Add to the main list
                all_org_repo.extend(repository_links)
            # Stop if we've found enough repositories
            if len(repositories) < 100:
                break
        except Exception as e:
            print(f"Error processing {org}: {str(e)}")
            continue

# Create the final DataFrame AFTER collecting all repositories
df = pd.DataFrame(all_org_repo, columns=['Organization']+prop_list)
print(f"\nTotal repositories fetched: {len(df)}")

# Save to CSV if needed
# df.to_csv("all_tech_companies_repos.csv", index=False)
```
